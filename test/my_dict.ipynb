```python
my_dict= {
    1: 123,
    12: 456,
    17: 789,
}
```

```python
# Find the nearest key in a dictionary with given value, while ensuring that the key is smaller than the given value
def find_nearest_key(my_dict, given_value):
    return my_dict[max(k for k in my_dict if k <= given_value)]
```

```python
find_nearest_key(my_dict= my_dict, given_value=11) #123
```

```python
from typing import Optional

def ensure_list_type(x) -> list:
    """
    Ensuring the outcome type is list, which as mentioned as the name of the function.
    """
    return (x if isinstance(x, list) else [x])


def _time_(
        value,
        subsidy_time: Optional[int] = None,
        **kwargs 
       ):
    time_combo = sorted(
        ensure_list_type(
            kwargs.get('prepay_arr', {}).get('time', [0]))
        + (ensure_list_type(subsidy_time) if subsidy_time else [])
    )
    return max(k if k <= value else 0 for k in time_combo )
```

```python
[_time_(value= n, prepay_arr={'time': [1, 6, 7, 8, 9]}, subsidy_time=13) for n in range(0, 361)]
```
